# Label Parameter

```python
days = [1,2,3,4,5,6,7]

max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.plot(days, max_t, label= "max")
plt.plot(days, min_t, label= "min")
plt.plot(days, avg_t, label= "average")

plt.legend(loc="center left", shadow = True, bbox_to_anchor= (1,.05))
```

# Bar Chart

```python
import numpy as np
```

```python
company = ["Google", "Amazon", "MS", "Facebook"]
revenue = [90, 136, 89, 27]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(xpos, revenue, label = "Revenue")
plt.ylabel("Revenue(Bin)")
plt.title("Technology Stocks")

plt.legend()
```

```python
profit = [40, 2,34,12]
```

```python
plt.bar(xpos-.2, revenue, width = .4, label = "Revenue")
plt.bar(xpos +.2, profit, width = .4, label ="Profit")

plt.xticks(xpos, company)
plt.ylabel("Revenue(Bin)")
plt.title("Technology Stocks")

plt.legend()
```
