## Pandas

```python
import pandas as pd
```

```python
df = pd.read_csv("loan_data.csv")
```

```python
df
```

```python
df.isna().sum()
```

## fillna

```python
df.fillna(0, inplace= True)
```

```python
df.fillna(
    {
        'purpose' :"mean"
    }

)
```

```python
df.fillna(method = "ffill")    #forward fill
```

```python
df.fillna(method= "bfill", inplace=True)     #backward fill
```

## Drop na

```python
df.dropna(inplace=True)
```

```python
df.dropna(how="all")     # when entire raw is na the drop it otherwise leave it
```

```python
df.dropna(thresh=2)         #drop if 2 columns in a raw is na 
```

## Replace

```python
import numpy as np
```

```python
df.replace(0, value=np.NaN)
```

```python
df.replace({
    "purpose" : 0,
    "fico" : "udsh"
}, np.NaN)
```
