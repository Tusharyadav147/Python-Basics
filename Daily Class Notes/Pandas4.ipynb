# Concatenation

```python
"A" + "B"
```

```python
indian_weather = pd.DataFrame({
    "city": ["mumbai", "delhi", "banglore"],
    "temperature": [32,45,60],
    "humidity": [80,60,78]
})

indian_weather
```

```python
nyc_weather = pd.DataFrame({
    "city": ["new york", "bsj", "banglore2"],
    "temperature": [567,546,65],
    "humidity": [80,60,78]
})

nyc_weather
```

```python
df = pd.concat([indian_weather, nyc_weather], ignore_index=True)
```

```python
df
```

```python
df = pd.concat([indian_weather, nyc_weather], keys= ["india", "us"])
df
```

```python
df.loc["india"]
```

```python
df.loc["us"]
```

```python
s = pd.Series(["Humid", "Dry", "Rain"], name = "Event")
```

```python
s
```

```python
df
```

```python
df = pd.concat([df, s], axis=1)
```
